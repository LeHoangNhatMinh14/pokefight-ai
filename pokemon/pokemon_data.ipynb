```python
import requests
import pandas as pd
import time

pokemon_list = []

for i in range(1, 721):
    url = f"https://pokeapi.co/api/v2/pokemon/{i}"
    response = requests.get(url)

    if response.status_code != 200:
        continue

    data = response.json()

    name = data["name"]

    types = [t["type"]["name"] for t in data["types"]]
    type1 = types[0]
    type2 = types[1] if len(types) > 1 else None

    stats = {stat["stat"]["name"]: stat["base_stat"] for stat in data["stats"]}

    pokemon_list.append({
        "name": name,
        "type1": type1,
        "type2": type2,
        "hp": stats.get("hp"),
        "attack": stats.get("attack"),
        "defense": stats.get("defense"),
        "sp_attack": stats.get("special-attack"),
        "sp_defense": stats.get("special-defense"),
        "speed": stats.get("speed"),
    })

    time.sleep(0.2)

df = pd.DataFrame(pokemon_list)

df.head()
```

```python
def assign_role(row):
    if row["speed"] >= 100 and (row["attack"] >= 90 or row["sp_attack"] >= 90):
        return "fast attacker"
    
    if row["attack"] >= 100:
        return "physical attacker"
    
    if row["sp_attack"] >= 100:
        return "special attacker"
    
    if row["hp"] >= 80 and (row["defense"] >= 80 or row["sp_defense"] >= 80):
        return "bulky defender"
    
    if row["defense"] >= 90 or row["sp_defense"] >= 90:
        return "defensive support"
    
    return "balanced"
```

```python
all_types = list(type_effectiveness.keys())

def get_type_multiplier(attacking_type, defender_type1, defender_type2=None):
    multiplier = type_effectiveness.get(attacking_type, {}).get(defender_type1, 1)
    
    if defender_type2 and not pd.isna(defender_type2):
        multiplier *= type_effectiveness.get(attacking_type, {}).get(defender_type2, 1)
    
    return multiplier


def get_defensive_profile(type1, type2=None):
    profile = {}
    
    for attacking_type in all_types:
        multiplier = get_type_multiplier(attacking_type, type1, type2)
        profile[attacking_type] = multiplier
    
    return profile


def get_real_weaknesses(type1, type2=None):
    profile = get_defensive_profile(type1, type2)
    return [attack_type for attack_type, multiplier in profile.items() if multiplier > 1]
```

```python
def recommend_teammates(selected_pokemon, df, playstyle="balanced", team_size=6, max_legendaries=0):
    selected_row = df[df["name"].str.lower() == selected_pokemon.lower()]
    
    if selected_row.empty:
        return f"{selected_pokemon} not found in dataset."
    
    # Allow both "offense" and "offensive"
    if playstyle == "offensive":
        playstyle = "offense"
    
    team = [selected_row.iloc[0].to_dict()]
    candidates = df[df["name"].str.lower() != selected_pokemon.lower()].copy()
    
    while len(team) < team_size:
        team_weaknesses = set()
        team_types = []
        team_roles = []
        
        for member in team:
            team_weaknesses.update(member["weaknesses"])
            team_roles.append(member["role"])
            
            team_types.append(member["type1"])
            
            if member["type2"] and not pd.isna(member["type2"]):
                team_types.append(member["type2"])
        
        # Count current legendaries
        legendary_count = sum(1 for member in team if member["is_legendary"])
        
        # Hard filter: if legendary limit reached, remove all legendary candidates
        if legendary_count >= max_legendaries:
            candidates = candidates[candidates["is_legendary"] == False]
        
        if candidates.empty:
            break
        
        def score_candidate(row):
            weaknesses = set(row["weaknesses"])
            weakness_overlap = len(team_weaknesses.intersection(weaknesses))
            
            type_duplication = 0
            
            if row["type1"] in team_types:
                type_duplication += 1
            
            if row["type2"] and not pd.isna(row["type2"]) and row["type2"] in team_types:
                type_duplication += 1
            
            role_duplication = team_roles.count(row["role"])
            
            atk = row["attack"]
            spa = row["sp_attack"]
            speed = row["speed"]
            hp = row["hp"]
            defense = row["defense"]
            spdef = row["sp_defense"]
            
            if playstyle == "offense":
                stat_score = (speed * 0.5) + (max(atk, spa) * 0.5)
            
            elif playstyle == "stall":
                stat_score = (hp * 0.4) + (defense * 0.3) + (spdef * 0.3)
            
            elif playstyle == "tank":
                stat_score = (max(atk, spa) * 0.5) + (hp * 0.3) + (defense * 0.2)
            
            else:  # balanced
                stat_score = (atk + spa + speed + hp + defense + spdef) / 6
            
            score = (
                weakness_overlap * 3
                + type_duplication * 2
                + role_duplication * 2
                - stat_score / 50
            )
            
            return score
        
        candidates["team_score"] = candidates.apply(score_candidate, axis=1)
        candidates = candidates.sort_values(by="team_score", ascending=True)
        
        next_pick = candidates.iloc[0]
        team.append(next_pick.to_dict())
        
        candidates = candidates[candidates["name"] != next_pick["name"]]
    
    return pd.DataFrame(team)
```

```python
import requests
import pandas as pd

def get_pokemon_moves(pokemon_name):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}"
    response = requests.get(url)

    if response.status_code != 200:
        return []

    data = response.json()
    moves = []

    for move_entry in data["moves"]:
        move_name = move_entry["move"]["name"]
        move_url = move_entry["move"]["url"]

        move_response = requests.get(move_url)

        if move_response.status_code != 200:
            continue

        move_data = move_response.json()

        # Ignore status moves
        if move_data["damage_class"]["name"] == "status":
            continue

        # Ignore moves with no power
        if move_data["power"] is None:
            continue

        moves.append({
            "move": move_name,
            "type": move_data["type"]["name"],
            "power": move_data["power"],
            "accuracy": move_data["accuracy"],
            "damage_class": move_data["damage_class"]["name"]
        })

    return moves
```

```python
def score_move(move):
    accuracy = move["accuracy"] if move["accuracy"] is not None else 100
    return move["power"] * (accuracy / 100)
```

```python
def recommend_moves(pokemon_name, df, max_moves=4):
    moves = get_pokemon_moves(pokemon_name)
    moves_df = pd.DataFrame(moves)

    if moves_df.empty:
        return pd.DataFrame()

    pokemon_row = df[df["name"].str.lower() == pokemon_name.lower()].iloc[0]
    pokemon_types = {pokemon_row["type1"]}

    if pokemon_row["type2"] and not pd.isna(pokemon_row["type2"]):
        pokemon_types.add(pokemon_row["type2"])

    moves_df["score"] = moves_df.apply(score_move, axis=1)
    moves_df["is_stab"] = moves_df["type"].apply(lambda move_type: move_type in pokemon_types)

    selected_moves = []
    used_types = set()

    # 1. Always pick the best STAB move first
    stab_moves = moves_df[moves_df["is_stab"]].sort_values(by="score", ascending=False)

    if not stab_moves.empty:
        best_stab = stab_moves.iloc[0]
        selected_moves.append(best_stab)
        used_types.add(best_stab["type"])

    # 2. Fill remaining slots with strong coverage moves
    coverage_moves = moves_df.sort_values(by="score", ascending=False)

    for _, move in coverage_moves.iterrows():
        if len(selected_moves) >= max_moves:
            break

        if move["move"] in [m["move"] for m in selected_moves]:
            continue

        if move["type"] not in used_types:
            selected_moves.append(move)
            used_types.add(move["type"])

    return pd.DataFrame(selected_moves)
```
